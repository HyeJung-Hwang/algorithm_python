# 스택과 큐의 구현

- 삽입 ( Push )

- 삭제 ( Pop )



# 1. 스택 ( Stack )

= 먼저 들어온 수는 가장 나중에 나간다.

= [ ], append(), pop() 로 구현

- 선입 후출 ( First in Last Out )

- 후입 선출 ( Last In First Out )

- 시간 복잡도 : O(1)


```python
stack = []

stack.append(5)
stack.append(3)
stack.append(2)
stack.append(7)

stack.pop()

stack.append(1)
stack.append(4)

stack.pop()

print(stack)
print(stack[::-1])
```

# 2. 큐 ( Queue )

= 먼저 들어온 수가 가장 먼저 나간다.

= collection 라이브러리의 deque, deque.append(),deque.popleft() 활용

- 선입선출 ( First In FIrst Out )


```python
from collections import deque

queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)

queue.popleft()

queue.append()

print(list(queue))
```

# 3. 재귀 함수



```python
def for_factorial(n):
    result = 1
    for i in range(1, n+1):
        result = result * i
    return result


def recursive_factorial(n):
    if n <= 1:
        return 1
    
    return result * for_factorial(n-1)

print(recursive_factorial(5))
```
