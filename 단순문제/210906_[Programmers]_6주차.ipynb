두 번째 시도: 다중 정렬 조건 세부적으로

```python
def solution(weights,head2head):
    def count_W(x):
        total=(x.count("W")+x.count("L"))
        if total==0:
            return 0
        else:
            return (1000*x.count("W"))/total
    number=[(i+1) for i in range(len(weights)) ]
    dic=dict(zip(weights,number))
    x=list(map(count_W,head2head))
    y=[0]*len(weights)
    for i in range(len(weights)):
        count=0
        for j in range(len(weights)):
            if head2head[i][j]=="W":
                if weights[i]<weights[j]:
                    count+=1
        y[i]=count
    z=weights
    w=number
    h=list(zip(x,y,z,w))
    s=sorted(h,key= lambda j: (-j[0],-j[1],-j[2],j[3]))
    for i in range(len(s)):
        s[i]=s[i][3]
            
    return s
```
