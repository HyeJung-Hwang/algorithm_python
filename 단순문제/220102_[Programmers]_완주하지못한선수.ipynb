파이썬 자료구조
- 해시 : 딕셔너리로 구현

```python
participant = ["mislav", "stanko", "mislav", "ana"]
```

```python
completion = ["stanko", "ana", "mislav"]
```

```python
def solution(participant,completion):
    a = {}
    for i in  participant:
        a[i] = a.get(i,0)+1
    #print(a)
    for i in completion:
        a[i] -= 1
    for i in a:
        if a[i]:
            return i
```
