첫번째 시도: for문 solution 함수 내에 구현

```python
n=3
```

```python
answer = 0
curr,prev=1,1
if n!=0 or n!=1:
    for i in range(n-2):
        prev,curr=curr,prev+curr   #curr=prev+curr(피보나치규칙)
        answer=curr
else:
    answer=n
answer%1234567
```

*피보나치 구현

```python
n=3
curr,prev=1,1
for i in range(n-2):
  prev,curr=curr,prev+curr #한번에 prev,curr 
  answer=curr
answer
```

```python
#잘못된 업데이트
n=4
curr,prev=1,1
for i in range(n-2):
  prev=curr
  curr=prev+curr   #이미 prev 값을 먼저 업데이트 해서 오류 
  answer=curr
answer
```
