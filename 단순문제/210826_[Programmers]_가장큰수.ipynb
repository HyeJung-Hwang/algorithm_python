첫 번째 시도: 8자리 2진수로 형식 맞춰서 구하기(오류)

```python
def solution(n):
    x=format(n,'b').zfill(8)
    y=x.count('1')
    z=x.find('1')
    answer=[0,0,0,0,0,0,0,0]
    answer[y]=1
    #answer=[0,0,0,1,0,0,0,0] 1뒤에부분 순열?
    answer="".join(answer)
    return answer
```

두 번째 시도: for 문 활용해서 "다음 큰 수 찾기"

```python
n=78
```

```python
x=format(n,'b').zfill(8)
y=x.count('1')
z=x.find('1')
```

```python
answer=["0","0","0","0","0","0","0","0"]
answer[z]="1"
#answer=[0,0,0,1,0,0,0,0] 1뒤에부분 순열?
answer="".join(answer)
```

```python
for num in (int("0b"+answer[z:],2),1000001):
    if str(bin(num)).count('1')==y:
        x=num
        break
```

세 번째 시도: 형식 안 맞추지 않고

```python
y=bin(n).count('1')
```
