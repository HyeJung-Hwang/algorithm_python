2차 시도 ( 220122, 성공 )

```python
def makeList(x):
    userReport = x.split()
    return (userReport[1] ,userReport[0] )   # [ 신고 당한, 신고한 ]

def solution(id_list, report, k):
    answer = []
    ans = {}
    ans1 = []
    y = list(set(list(map(makeList,report))))
    
    
    fiDict = {}
    d = {}
    for i in id_list:
	    d.setdefault(i,0)
    for i in range(len(y)):
       if y[i][0] in fiDict.keys():
           fiDict[y[i][0]].append(y[i][1])
       else:
           fiDict[y[i][0]] = [y[i][1]]
    z = list(fiDict.values())
   
    for i in range(len(z)):
        if len(z[i]) >= int(k):
            for j in z[i]:
                d[j] += 1
    return list(d.values())
```

```python
solution(["muzi", "frodo", "apeach", "neo"]	,["muzi frodo","apeach frodo","frodo neo","muzi neo","apeach muzi"],	2) #	[2,1,1,0]
```
