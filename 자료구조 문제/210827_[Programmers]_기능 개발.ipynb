```python
progresses=[93,30,55]
speeds=[1,30,5]
```

queue FIFO: first in 까지 코드 내에 구현한 것은 아니지만 first out 구현( 앞에 있는 요소 부터 out )

```python
answer=[]
time=0
count=0
```

```python
while len(progresses)>0:  # out 이 진행되는 queue 의 길이로 while문 조건
  if (progresses[0]+time*speeds[0]) >= 100:
    progresses.pop(0)
    speeds.pop(0)
    count+=1
  else:
    if count>0:
      answer.append(count)  #answer 채우는 코드
      count=0
    time+=1
answer.append(count)
```
