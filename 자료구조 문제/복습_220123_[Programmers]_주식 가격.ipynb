정답 코드

```python
def solution(prices):
    answer = [0]*len(prices)
    stack = []
    for time, item in enumerate(prices):
        while stack and stack[-1][1] > item:
            v = stack.pop()
            answer[v[0]] = time- v[0] 
        
        stack.append([time, item])
    
    # 끝까지 가격이 떨어지지 않은 애들
    while stack:
        time = stack.pop()[0]
        answer[time] = len(prices)-1 - time
        
    return answer
```
