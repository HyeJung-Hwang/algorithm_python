첫번째시도: 시간초과

```python
prices=[1,2,3,2,3]
```

```python
stk=[]
count=0
```

```python
for j in range(len(prices)):
  x=prices[j]
  for i in prices[j+1:]:
    count+=1
    if i<x:
      break
  stk.append(count)
  count=0
```

두번째 시도: 스택 활용

```python
prices=[1,2,3,2,3]
```

```python
stack=[] #index들 집어 넣기
```

```python
answer=[0]*len(prices)
stack=[]
```

주식 가격이 떨어지는 경우

```python
for i, price in enumerate(prices):
  while stack and price< prices[stack[-1]]:
    j=stack.pop()
    answer[j]=i-j #가격이 떨어졌을 때
  stack.append(i) #전값 기록
```

```python
stack
```

```python
answer
```

주식가격이 떨어지지 않는 경우

```python
while stack:
  j=stack.pop()
  answer[j]=len(prices)-1-j
```

```python
answer
```
