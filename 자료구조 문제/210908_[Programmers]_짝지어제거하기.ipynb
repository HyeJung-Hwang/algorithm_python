스택 활용하기

```python
def solution(s):
    # 문자열의 길이가 홀수일 때
    if len(s) % 2 == 1: return 0 
    # 문자열의 길이가 2일 때
    if len(s) == 2: 
        return 1 if s[0] == s[1] else 0
    # 문자열의 길이가 2가 아닌 짝수 일 때        
    stack = [s[0]]
    
    for v in s[1:]:
        if len(stack) > 0 and stack[-1] == v:
            stack.pop()
        else:
            stack.append(v)
            
    return 0 if len(stack) else 1
```
