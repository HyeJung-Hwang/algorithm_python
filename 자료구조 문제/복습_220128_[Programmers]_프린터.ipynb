```python
def solution(priorities, location):
    
    que = priorities
    que5 = list(range(len(priorities)))
    que6 = []
    que3 = []
    while True:
        try:
            if que[0] < max(que[1:]):
                que.append(que.pop(0))
                que5.append(que5.pop(0))
            else:
                que3.append(que.pop(0))
                que6.append(que5.pop(0))

        except:
            break        

    #print(que3,que)
    ans = que6 + que5
    print(ans.index(location) + 1)
    #return 
    #return que2.index(location)+1
```
