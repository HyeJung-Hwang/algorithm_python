Greedy 알고리즘

```python
number="1924"
k=2
```

stack 리스트로 구현

stk.pop()


*   리스트 뒤에서 빼기


stk.append(요소)


*   리스트 뒤에서 넣기



```python
answer = ''
stk = []
for i in number:
    while stk and stk[-1] < i and k>0:
        k-=1
        stk.pop()
    stk.append(i)
```
