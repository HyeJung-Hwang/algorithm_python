Greedy: " 가장 큰 순서대로 " ,"가장 작은 순서대로 " 라는 표현인 있는 문제에서 사용
-정렬문제와 관련

```python
n = 1260
count = 0
#Greedy
coin_types = [500, 100, 50, 10] # 큰 단위의 화폐부터 차례대로 확인하기  
```

```python
for coin in coin_types:
    count += n // coin # 해당 화폐로 거슬러 줄 수 있는 동전의 개수 세기
    n %= coin   #처음 금액에서 돈 주고 남은 금액으로 update
```

```python
print(count)
```
