# 거스름돈 문제

->1260원을 최소의 동전 개수로 남는 돈 없이 거슬러주고 쓰인 동전 개수 구하기

```python
n = 1260
coin_array = [500,100,50,10]
count = 0

for i in coin_array:
    
    count += n // i
    n = n % i
```

# 숫자 카드 게임
-> 여러 개의 숫자 카드 중 가장 높은 숫자가 쓰인 카드 1 장을 뽑는 게임

-> 카드가 NXM 형태로 놓여있음

-> 먼저 행 선택 후 가장 숫자

```python
import sys

n,m = sys.stdin.readline().split()
num_list = []
for i in range(int(n)):
    temp = min(list(map ( int, sys.stdin.readline().split() )))
    num_list.append(temp)

print(max(num_list))
```

# 1이 될 때까지

-> 어떠한 수 N이 1이 될 때까지, [ 1. N에서 1 빼기 ], [ 2. N을 K로 나눈다 ( N이 K로 나누어 떨어질 때만] 중 선택해서 수행 가능

-> 수행의 최소 개수 구하기

```python
n, k = map(int, input().split())
cnt = 0

while n > 1:

    if n % k == 0:
        n = n // k
        cnt += 1

    else:
        n -= 1
        cnt += 1

print(cnt)
```
