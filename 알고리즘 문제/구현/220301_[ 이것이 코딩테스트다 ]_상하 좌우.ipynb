```python
import sys
```

```python
n = int(input())
plan = sys.stdin.readline().split()
row, column = 1,1

for d in plan: 
    if d == 'L':
        if column != 1:
            column -= 1
  # 문자가 R이라면 현재 열이 N이라면 무시, 아니면 열+1    
    elif d == 'R':
        if column != n:
            column += 1
  # 문자가 U라면 현재 행이 1이라면 무시, 아니면 행-1
    elif d == 'U':
        if row != 1:
            row -= 1
  # 문자가 D라면 현재 행이 N이라면 무시, 아니면 행+1
    elif d == 'D':
        if row != n:
            row += 1
```
