# 상하 좌우

-> 시뮬레이션 문제

```python
n = int(input())
plan = sys.stdin.readline().split()
row, column = 1,1

for d in plan: 
    if d == 'L':
        if column != 1:
            column -= 1
  # 문자가 R이라면 현재 열이 N이라면 무시, 아니면 열+1    
    elif d == 'R':
        if column != n:
            column += 1
  # 문자가 U라면 현재 행이 1이라면 무시, 아니면 행-1
    elif d == 'U':
        if row != 1:
            row -= 1
  # 문자가 D라면 현재 행이 N이라면 무시, 아니면 행+1
    elif d == 'D':
        if row != n:
            row += 1
```

# 브루트 포스

-> 모든 경우의 수 검증

```python
# hour : 00 ~ N ( N < 24 )
# minute : 00 ~ 59 
# sec : 00 ~ 59
 

 n = int(input())
 count = 0

 for h in range(n+1):
     for m in range(60):
         for s in range(60):
             if '3' in str(h) + str(m) + str(s):
                count = count + 1

print(count)
```
